```python
import pylab as pl
%pylab inline
import numpy as np
import os
import glob
import time
from mlabwrap import mlab
```

```python
class ImageUtils(object):
    """ Basic Image Processing Utils for Reid
    This class contains several basic operation functions on patch, a set of patches, single image
    *. get a patch from an image
    *. visualize a set of patches
    *. single image processing
    *. and so on ...
    """
    def __init__(self, image):
        """ Initialize the parameters of image utils
            : type image: numpy.ndarray
            : param image: an image with single channel or multiple channels
        """
        self.image = image
        self.h, self.w = image.shape[0:2]
        
        
    def subwin(self, rect):
        """ crop a sub window in an image
        """
        mlab.addpath('/home/rzhao/Projects/matlab-reid/imgproc/', nout=0)
        subim = mlab.subwin(self.image, rect)
        return subim
```

```python
class DatasetUtils(object):
    """ Dataset Processing Utils for Reid

    This library contains a lot of basic operation functions mostly used in reid task
    *. load images
    *. load labels
    *. load other auxilary data
    *. generate partitions
    *. load partitions
    *. visualize a ranking of images
    *. visualize a set of images
    """
    def __init__(self, dataset):
        """ Initialize the parameters of dataset utils
            :type dataset: string
            :param dataset: dataset name, and through telling the what dataset is, 
                            we customize some dataset-specific functions
        """
        self.dataset = dataset
        
        if dataset == 'viper':
            self.dataset_dir = '/home/rzhao/Projects/data/viper/'
            self.image_ext = 'png'
        elif dataset == 'campus':
            self.dataset_dir = '/home/rzhao/Projects/data/'
    
    def load_images(self):
        files = sorted(glob.glob(self.dataset_dir + '*.' + self.image_ext))
        self.h, self.w = pl.imread(files[0]).shape[0:2]
        return files
    
    def get_densefeature(self, files):
        """ matlab wraper for extracting dense color sift features
        """
        mlab.addpath('/home/rzhao/Projects/matlab-reid/dense_feat/', nout=0)
        self.gridstep, self.patchsize = 4, 10
        self.dim = 672
        self.nx = len(xrange(self.patchsize/2, self.w-self.patchsize/2+1, self.gridstep))
        self.ny = len(xrange(self.patchsize/2, self.h-self.patchsize/2+1, self.gridstep))
        features = np.empty((self.dim, self.ny*self.nx, len(files)))
        for i in xrange(0, len(files)): 
            tic = time.clock()
            features[:, :, i] = mlab.get_densefeature(plt.imread(files[i]))
            toc = time.clock()
            print 'Extracting dense features for image {} [time cost {}]'.format(i, toc-tic)
        
        return features
        
```
